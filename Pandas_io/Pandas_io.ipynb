# 1 - TRABALHANDO COM ARQUIVOS CSV

## 1.1 - Leitura de Arquivo CSV

Arquivos CSV são padrões muito comuns quando transferimos ou exportamos dados de diferentes ferramentas, por seu formato de fácil leitura. Então é algo que vemos com frequência.

```python
path_superstore_data = 'in/superstore_data.csv'
superstore_data = pd.read_csv(path_superstore_data)

superstore_data.head()
```

## 1.2 - Parâmetros Adicionais na Leitura de CSV

### SEP

Podemos usar o parâmetro `sep=` para definir o separador dos dados, isso é útil quando o separador não segue o padrão **(que é ",")**.

```python
path_superstore_data_ponto_virgula = 'in/superstore_data_ponto_virgula.csv'

superstore_data_ponto_virgula = pd.read_csv(path_superstore_data_ponto_virgula, sep=';')
superstore_data_ponto_virgula.head()
```

### NROWS

Quando temos um dataset muito grande e queremos analisar apenas uma parte dele podemos usar o parâmetro `nrows=`. Com ele definimos a quantidade de linhas que devem ser lidas.

```python
primeiros_registros_superstore_data = pd.read_csv(path_superstore_data, nrows=5)
primeiros_registros_superstore_data
```

### USECOLS

De forma semelhante podemos usar o `usecols=` para selecionar quais colunas queremos ler em um dataset. Sendo que para isso podemos usar os índices ou nomes das colunas.

```python
selecao_por_nome = ['Id', 'Year_Birth', 'Income']
selecao_por_nome_superstore_data = pd.read_csv(path_superstore_data, usecols=selecao_por_nome)

selecao_por_nome_superstore_data
```

```python
selecao_por_posicao = [0, 1, 4] # Mesmo que as colunas Id, Year_Birth, Income
selecao_por_posicao_superstore_data = pd.read_csv(path_superstore_data, usecols=selecao_por_posicao)

selecao_por_posicao_superstore_data
```

## 1.3 - Escrevendo Arquivos CSV

Como é um formato muito utilizado é normal precisarmos salvar esse tipo de arquivo.

Sendo que é normal usarmos o `sep=` aqui também para mudar o separador, além disso se usarmos o `index=False` removemos o index, fazendo com que não seja salvo no arquivo (muitas vezes o recomendado).

```python
path_out = 'out/clients_superstore.csv'
selecao_por_nome_superstore_data.to_csv(path_out, sep=';')
```

```python
clients_superstore = pd.read_csv(path_out, sep=';')
clients_superstore
```

```python
selecao_por_nome_superstore_data.to_csv(path_out, sep=';', index=False)

pd.read_csv(path_out, sep=';')
```

# 3 - LIDANDO COM JSON

## 3.1 - Leitura de JSON Simples

```python
pacientes = pd.read_json('in/pacientes.json')
pacientes.head()
```

Porém o mais comum em arquivos JSON é termos estruturas aninhadas, nesses casos precisamos fazer um tratamento particular para normalizar os dados.

```python
pacientes_2 = pd.read_json('in/pacientes_2.json')
pacientes_2
```

## 3.2 - Normalizando arquivos JSON (Lendo Estruturas Aninhadas)

```python
pd.json_normalize(pacientes_2['Pacientes'])
```

```python
try:
    pd.json_normalize(pacientes_2, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
except:
    print('Erro ao normalizar arquivo.')
```

Se tentarmos fazer a normalização de um arquivo JSON diretamente isso pode gerar erros, para isso precisamos usar a lib json para que o arquivo já seja tratado diretamente e tenhamos um tratamento mais simples! Sendo que ao usarmos o `meta` mantemos as informações mesmo normalizando os dados, isso é interessante para não perdermos informações na normalização.

```python
#Importando o módulo JSON
import json

#Lendo o arquivo json usando o módulo Python JSON
with open('in/pacientes_2.json','r') as f:
    dados = json.loads(f.read())

#Normalizando os dados com os parâmetros record_path e meta
pacientes_2_normalizado = pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
pacientes_2_normalizado
```

## 3.3 - Salvando em JSON

```python
pacientes_2_normalizado.to_json('out/pacientes_2_normalizado.json')
pd.read_json('out/pacientes_2_normalizado.json')
```

# 5 - CONSUMINDO BANCO DE DADOS (SQL)

## 5.1 - Setup Inicial Banco de Dados

Para podermos fazer os testes em ambiente local usaremos um banco sqlite em memória. Com isso poderemos fazer os comandos e criar um banco pronto para uso.

```python
import sqlalchemy

sqlalchemy.__version__
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
con = create_engine('sqlite:///:memory:')
```

## 5.2 - Criando Tabelas no Banco SQL

Agora que temos uma conexão com o banco de dados podemos usar essa conexão para adicionar dados em uma nova tabela desse banco. Isso pode ser feito usando o comando `to_sql(<table_name>, con=<connection>)`, então teremos os dados salvos como uma nova tabela (se o mesmo nome já não estiver em uso). 

```python
clientes_banco = pd.read_csv('in/clientes_banco.csv')
clientes_banco.head()
```

```python
clientes_banco.to_sql('clientes_banco', con, index=False, if_exists='replace')
```

Uma vez que temos a tabela criada no banco podemos verificar se ela existe usando `inspect(<connection>)` e depois `.get_table_names()` para verificar quais tabelas existem no banco. 

```python
inspector = inspect(con)
inspector.get_table_names()
```

## 5.3 - Leitura com Consultas SQL e Leitura de Tabelas

Quando usamos o Pandas podemos ler dados de bancos de dados de duas formas distintas, lendo queries SQL ou tabelas diretamente.

Ao lermos uma consulta SQL usamos a consulta para coletar os dados expecíficos e usar esse retorno como um DataFrame. Para isso usamos o comando `read_sql`.

Por outro lado, quando queremos diretamente toda a tabela podemos usar um formato mais simples e pegar diretamente a tabela com `read_sql_table`.

**Em ambos os casos usamos como base a connection criada com o banco de dados.**

```python
query = 'SELECT * FROM clientes_banco WHERE Categoria_de_renda = "Empregado"'

# Categoria_de_renda foi removido por so ter 'Empregado'
empregados = pd.read_sql(query, con).drop('Categoria_de_renda', axis=1)
empregados.head()
```

```python
empregados.to_sql('empregados', con, index=False, if_exists='replace')
inspector.get_table_names()
```

```python
pd.read_sql_table('empregados', con)
```

```python
pd.read_sql_table('empregados', con, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

## 5.4 - Atualizando Registros do Banco de Dados

```python
query = 'SELECT * FROM clientes_banco'

pd.read_sql(query, con)
```

### Removendo Registros

Podemos usar a clausula `DELETE FROM` para remover registros do banco de dados. Para isso só precisamos usar a conection que já temos e executar uma query como uma consulta, dessa forma teremos o código executado e o banco modificado.

```python
query = 'DELETE FROM clientes_banco WHERE ID_Cliente = 5008804'

with con.connect() as conn:
    conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes_banco', con)
```

### Atualizando Registros

Podemos usar a clausula `UPDATE <table> SET` para atualizar registros de uma tabela. Utilizaremos para isso o mesmo formato usado para remover um registro.

```python
query = 'UPDATE clientes_banco SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'

with con.connect() as conn:
    conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('clientes_banco', con)
```
