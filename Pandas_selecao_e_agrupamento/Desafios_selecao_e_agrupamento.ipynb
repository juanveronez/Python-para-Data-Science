```python
import pandas as pd
```

```python
path = 'in/1-SEEG10_GERAL-BR_UF_2022.10.27-FINAL-SITE.xlsx'
sheet_name = 'GEE Estados' # Para os desafios tambem usaremos a tabela de GEE Estados!

gee_estados = pd.read_excel(path, sheet_name)

gee_estados = gee_estados[gee_estados['Emissão / Remoção / Bunker'] == 'Emissão']

gee_estados.head()
```

```python
# 1) Encontre os valores únicos das colunas "Nível 1 - Setor" e "Estado" para identificar
# as atividades econômicas presentes na base de dados e se todos os Estados do Brasil
# estão presentes no DataFrame.

print('Setores:', gee_estados['Nível 1 - Setor'].unique(),'\n')
print('Estados:', gee_estados['Estado'].unique())
```

```python
# 2) Filtre o DataFrame somente com os dados dos Estados da região Sul do Brasil.

estados_sul = ['RS', 'SC', 'PR']
gee_estados_sul = gee_estados[gee_estados['Estado'].isin(estados_sul)]

gee_estados_sul.head()
```

```python
# 3) Filtre o DataFrame somente com os dados de "Mudança de Uso da Terra e Floresta" que sejam do Estado do Amazonas.
filtro = (gee_estados['Nível 1 - Setor'] == 'Mudança de Uso da Terra e Floresta') & (gee_estados['Estado'] == 'AM')

gee_estados[filtro].head()
```

```python
# 4) Encontre o valor máximo de emissão do ano de 2021 para os dados de "Agropecuária" no Estado do Pará.

filtro = (gee_estados['Nível 1 - Setor'] == 'Agropecuária') & (gee_estados['Estado'] == 'PA')
gee_estados.loc[filtro, 2021].max()
```

1) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para visualizar o dicionário contendo as chaves de grupos formados e a lista de índices de cada grupo.

2) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" e localize os dados do grupo "Agropecuária".

3) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para identificar a média de emissão de cada atividade econômica no ano de 2021.

## Setup módulo 2

```python
colunas_info = gee_estados.loc[:, :'Produto'].columns.to_list()
colunas_emissao = gee_estados.loc[:, 1970:].columns.to_list()

wide_gee_estados = gee_estados

gee_estados = gee_estados.melt(colunas_info, colunas_emissao, 'Ano', 'Emissão')
```

```python
gee_estados.head()
```

## Exercícios Módulo 2



```python
# 1) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para visualizar
# o dicionário contendo as chaves de grupos formados e a lista de índices de cada grupo.

gee_estados_por_setor = gee_estados.groupby('Nível 1 - Setor')
gee_estados_por_setor.groups
```

```python
# 2) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor"
# e localize os dados do grupo "Agropecuária".

gee_estados_por_setor.get_group('Agropecuária').head()
```

```python
# 3) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para
# identificar a média de emissão de cada atividade econômica no ano de 2021.

gee_estados[gee_estados['Ano'] == 2021] \
    .groupby('Nível 1 - Setor')[['Emissão']].mean() \
    .sort_values('Emissão', ascending=False)
```

```python
# 4) Faça um agrupamento de dados com base na coluna "Nível 1 - Setor" para
# identificar a soma de emissão de cada atividade econômica. Ordene os dados
# da maior para menor emissão.

gee_estados_por_setor[['Emissão']].sum().sort_values('Emissão', ascending=False)
```

```python
# 1) Faça um agrupamento de dados com as colunas "Estado" e "Nível 1 - Setor", obtendo a soma de emissão e armazenando o resultado em uma tabela.

gee_estados.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
```

```python
gee_estado_setor = gee_estados.groupby(['Estado', 'Nível 1 - Setor'])[['Emissão']].sum()
gee_estado_setor
```

```python
# 2) Utilizando a tabela construída na atividade 1, selecione os dados referentes à "Energia" do índice "Nível 1 - Setor".

gee_estado_setor.swaplevel(0, 1).loc['Energia']
# Versao melhor para a mesma coisa \/
gee_estado_setor.xs('Energia', level = 1)
```

```python
# 3) Utilizando a tabela construída na atividade 1, encontre a atividade econômica com valor máximo de emissão do Estado de Minas Gerais

gee_estado_setor.loc['MG'].agg(['idxmax', 'max'])
```

```python
# 4) Obtenha uma tabela contendo a atividade econômica com máxima emissão para cada Estado.

gee_estado_setor.groupby(level=0).idxmax()
```

```python
# 5) Obtenha uma tabela contendo o Estado com máxima emissão para cada atividade econômica.

gee_estado_setor.groupby(level=1).idxmax()
```
