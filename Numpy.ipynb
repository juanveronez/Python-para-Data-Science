## 2.3 Lidando com NaNs

Valores nulos são comuns em arrays, seja por problemas no armazenamento, coleta ou tratamento de dados. Entretando é importante encontrar esses valores e lidar com esses casos.

Nesse momento veremos como identificar um valor nulo (visualmente e através de códigos) e tanbém como alterar um valor nulo.

```python
plt.plot(datas, Kaliningrad)
```

```python
print(f'Quantidade de NaNs no Array: {np.isnan(Kaliningrad).sum()}')
```

Uma vez que sabemos que existem valor nulos (tanto visualmente como por código) podemos substituir essa ocorrência. Nesse caso o que faremos é fazer com que o valor seja igual à média dos vizinhos.

```python
def substitui_nans(array= np.array([])):
    res_array = array.copy()
    shape = np.arange(0, array.shape[0])
    
    for line, i in zip(array, shape):
        if np.isnan(line):
            res_array[i] = (res_array[i-1] + res_array[i+1])/2
    
    return res_array
```

```python
if np.isnan(Kaliningrad).any():
    Kaliningrad = substitui_nans(Kaliningrad)

plt.plot(datas, substitui_nans(Kaliningrad))
```

A média também pode ser usada para comparar dois arrays, sendo uma comparação de uma medida de sumarização dos arrays. **Isso não seria possível caso tivéssemos valores núlos (o cálculo de média não funciona com NaNs)**

```python
print(f'Média de preço em Moscow: {np.round(np.mean(Moscow), 2)}')
print(f'Média de preço em Kaliningrad: {np.round(np.mean(Kaliningrad), 2)}')
```

# 3. Operações Entre Arrays

## 3.1 Diferença entre Arrays

Para entendermos a tendência de uma função, no caso os dados que estamos analisando, uma ferramenta muito útil é criarmos uma **linha de tendência**. Uma linha de tendência é criada como uma função linear que mais se ajusta aos dados apresentados. Sendo que quando falamos "mais se ajustam" queremos dizer que o valor da diferença entre ambas é minimizado.

Apenas para relembrar, uma função linear é: `y = a*x + b`

```python
y = 2*datas + 80 # Valores meramente colocados sem nenhum critério

plt.plot(datas, Moscow)
plt.plot(datas, y)

plt.show()
```

Uma vez que queremos entender de forma não visual o quão distante duas funções estão precisamos compará-las, isso pode ser feito comparando suas distâncias, como no cálculo abaixo

```python
def func_erro(y1, y2):
    return np.sqrt(((y1-y2)**2).sum())
```

```python
y = 0.43*datas + 80 # Obtido testando valores com a função abaixo

plt.plot(datas, Moscow)
plt.plot(datas, y)
plt.title(f'Erro da função: {func_erro(Moscow,y):.2f}')
plt.show()
```

```python
print(f'Norma da comparação entre arrays: {np.linalg.norm(Moscow-y)}')
if np.equal(np.linalg.norm(Moscow-y), func_erro(Moscow, y)):
    print('\nA função de norma do numpy é igual a função criada anteriormente!')
```

## 3.2 Cálculo de Regressão Linear

Uma vez que queremos criar uma função que se aproxime de uma data função utilizando uma função linear o que estamos criando é uma função linear.

Sendo que essa função tem como preceito se aproximar da tendência da função original, mesmo com uma taxa de erro.

Nesses casos o que utilizamos é a **Regressão Linear**.

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

O coeficiente angular é o angulo de crescimento de X no eixo Y, seu cálculo é dado por:


```python
a = (n*(X*Y).sum() - X.sum()*Y.sum()) / (n*((X**2).sum())-(X.sum())**2)
a
```

Já o coeficiente linear representa em que ponto a função corta o eixo X quando Y=0



```python
b = Y.mean() - a*X.mean()
b
```

```python
y = a*X + b

print(f'Norma da função: {np.linalg.norm(Moscow-y)}')
```

## 3.3 Usos de Regressão Linear

A regressão linear nos ajuda á encontrar uma função aproximada que pode representar uma dada função. Sendo usada para encontrar valores no futuro, passado ou até mesmo intermediários.

É importante notar que essa função se trata de uma aproximação simples, então não é esperado que o valor seja aquele, mas sim um valor apoximado com uma taxa de erro.

```python
plt.plot(datas, Moscow)
plt.plot(datas, y)
plt.plot(41.5,41.5*a+b, '*y')
plt.plot(100,100*a+b, '*')
plt.plot(-5,-5*a+b, '*g')

plt.show()
```

Então com esse tipo de cálculo fazemos algo muito valioso, que é estimar com um certo erro qual seria um valor em um dado ponto, mesmo sem ter essa informação exata.

Utilizando sementes para preservar resultados aleatórios, dessa forma sempre que a função for usada teremos o mesmo resultado.

```python
def rand_regressao_linear(lim_a, lim_b, X):
    np.random.seed(42)
    rand_a = np.random.uniform(low=lim_a['low'], high=lim_a['high'], size=100)
    rand_b = np.random.randint(low=lim_b['low'], high=lim_b['high'], size=100)
    rand_norm = np.array([])
    rand_reg = {
        'norm': np.Infinity
    }

    for a, b in zip(rand_a, rand_b):
        norm = np.linalg.norm(Moscow-(a*X+b))
        rand_norm = np.append(rand_norm, norm)
        if norm < rand_reg['norm']:
            rand_reg = { 'norm': norm, 'a': a, 'b': b }

    return rand_reg, { 'rand_a': rand_a, 'rand_b': rand_b, 'rand_norm': rand_norm }


melhor_rand, rands = rand_regressao_linear({ 'low': 0.1, 'high': 0.9 }, { 'low': 40, 'high': 100 }, X)
```
